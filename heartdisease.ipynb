Robust Scaler
This method removes the median and scales the data in the range between 1st quartile and 3rd quartile.


If outliers are present in the dataset, then the median and the interquartile range provide better results and outperform the sample mean and variance. RobustScaler uses the interquartile range so that it is robust to outliers

```python
x = pd.DataFrame({
    # Distribution with lower outliers
    'x1': np.concatenate([np.random.normal(20, 2, 1000), np.random.normal(1, 2, 25)]),
    # Distribution with higher outliers
    'x2': np.concatenate([np.random.normal(30, 2, 1000), np.random.normal(50, 2, 25)]),
})

x.head()
```

"""
np.concatenate == dizileri 2 boyutlu eksen hale getirir satır
np.random.normal ise  normal dağılımla sahip bir dizi oluşturu
ilk index mean yani ortalama
ikinci index standart sapma
3.index kaç değer alacağı 
"""

```python
scaler = preprocessing.RobustScaler()
robust_df = scaler.fit_transform(x)
robust_df = pd.DataFrame(robust_df,columns=["x1","x2"])
robust_df.head()
```

```python
scaler = preprocessing.StandardScaler()
standard_df = scaler.fit_transform(x)
standard_df = pd.DataFrame(standard_df, columns =['x1', 'x2'])
standard_df.head()
```

```python
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(x)
minmax_df = pd.DataFrame(minmax_df, columns =['x1', 'x2'])
minmax_df.head()
```
